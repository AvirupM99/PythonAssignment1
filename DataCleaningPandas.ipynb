```python
#DataCleaningPandas

import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
df
```

```python
#Q1
for i in range(1, df['FlightNumber'].count() + 1):
    if pd.isnull(df.loc[i, 'FlightNumber']):
        df.loc[i, 'FlightNumber'] = df.loc[i-1, 'FlightNumber'] + 10
df['FlightNumber']
df
```

```python
df['FlightNumber'] = df['FlightNumber'].dropna().astype(int)
df
```

```python
display(df.dtypes)
```

```python
#Q2
copyDf = df.copy()
copyDf[['From', 'To']] = copyDf.From_To.str.split('_', expand=True)
copyDf
```

```python
#Q3
copyDf.From = copyDf.From.str.capitalize()
copyDf.To = copyDf.To.str.capitalize()
copyDf.From_to = copyDf.From_To.str.capitalize()
print(copyDf)
```

```python
#Q4
copyDf = copyDf.drop(copyDf.columns[0], axis=1)
copyDf
```

```python
copyDf = copyDf[['Airline','FlightNumber', 'RecentDelays', 'From', 'To',]]
copyDf
```

```python
#Q5
copyDf_delays=pd.DataFrame(copyDf["RecentDelays"].values.tolist())
copyDf_delays.columns = ['delay_{}'.format(n) for n in range(1, len(copyDf_delays.columns)+1)]
copyDf_delays
```

```python
copyDf = copyDf.drop("RecentDelays" , axis=1).join(copyDf_delays)
df = copyDf
df
```

```python
#FinalOutput
newIndex = np.arange(1, df.Airline.count() + 1)
newIndex 
df.set_index(newIndex, inplace=True)
df
```
